```python
import datetime
from dateutil.parser import parse

# Unify the date format

def unify(string):
    final = "".join(string)
    final = parse(final, default=datetime.datetime(2015, 1, 1)).strftime("%Y%m%d")
    return final

df['unidates'] = df['dates'].apply(unify)
df = df.sort_values('unidates')
df
```

```python
# Make dates more visible

def visible (string):
    final = string[:4] + '/' + string[4:6] + '/' + string[6:]
    return final

df['visible'] = df['unidates'].apply(visible)

df
```

```python
# Create "decades" categories for potential text classification

def decades (string):
    final = string[:3]
    return final

df['decades'] = df['unidates'].apply(decades)

df
```
